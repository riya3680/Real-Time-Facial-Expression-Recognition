```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout
from keras.layers import Convolution2D, MaxPooling2D, ZeroPadding2D
from keras.optimizers import SGD
import cv2
from keras.utils import to_categorical
from keras_preprocessing.image import load_img
import os
import tensorflow as tf
```

```python
TRAIN_DIR = 'images\train'
TEST_DIR = 'images\train'
```

```python
print(TRAIN_DIR)
```

```python
print(TEST_DIR)
```

```python
import os
def createdataframe(datadir):
    image_paths=[]
    labels=[]
    for label in os.listdir(datadir):
        print(label)
        for imagename in os.listdir(os.path.join(datadir,label)):
            image_paths.append(os.path.join(datadir,label,imagename))
            labels.append(label)
        print(label, "completed")
        
    return image_paths,labels
```

```python
train = pd.DataFrame()
train['image'], train['label'] = createdataframe(TRAIN_DIR)
```

```python
print(train)
```

```python
test= pd.DataFrame()
test['image'], test['label'] = createdataframe(TEST_DIR)
```

```python
print(TEST_DIR)
```

```python
from tqdm.notebook import tqdm
```

```python
def extract_features(images):
    features=[]
    for image in tqdm(images):
        img = load_img(image,grayscale=True)
        img = np.array(img)
        features.append(img)
    features = np.array(features)
    features = features.reshape(len(features),48,48,1)
    return features
```

```python
train_features = extract_features(train['image'])
```

```python
test_features = extract_features(test['image'])
```

```python
x_train=train_features/255.0
x_test=test_features/255.0
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
! pip install -U scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(train['label'])
```

```python
y_train = le.transform(train['label'])
y_test = le.transform(test['label'])
```

```python
y_train  = to_categorical(y_train,num_classes=7)
y_test  = to_categorical(y_test,num_classes=7)
```

```python
model = Sequential()

model.add( Convolution2D(128,kernel_size=(3,3),activation='relu',input_shape=(48,48,1)))
model.add(Dropout(0.4))

model.add( Convolution2D(256,kernel_size=(3,3),activation='relu'))
model.add(Dropout(0.4))

model.add(Flatten())
#fully connected layers
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))

#output layer
model.add(Dense(7,activation='softmax'))

```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy()
              ,metrics=['accuracy'])
```

```python
logdir='images'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(x=x_train,y=y_train ,epochs=2, batch_size=128,validation_data=(x_test, y_test),callbacks=[tensorboard_callback])
```

```python
print(hist)
```
